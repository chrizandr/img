```python
# Import functions to use in the code.

# Functions to read the image and convert to 8bit representation.
from skimage.io import imread
from skimage.transform import resize
import numpy as np
import time

# Maplotlib to plot the images.
import matplotlib.pyplot as plt
%matplotlib inline

# Main functions for finding and showing dominant colors
from Q9 import mean_filter, fast_mean_filter, median_filter, fast_median_filter
```

#### Mean Filtering:

```python
img = imread('A2_resources/barbara.jpg', as_gray=True)

start = time.time()
mean = mean_filter(img, 3)
end = time.time()

print("Filtered in {} seconds".format(end-start))

start = time.time()
mean_fast = fast_mean_filter(img, 3)
end = time.time()

print("Filtered in {} seconds".format(end-start))

f, ax = plt.subplots(1,3, figsize=(15,15))
ax[0].imshow(img, 'gray')
ax[0].set_title("Original Image")

ax[1].imshow(mean, 'gray')
ax[1].set_title("Mean filtered Image")

ax[2].imshow(mean_fast, 'gray')
ax[2].set_title("Fast Mean filtered image")

plt.show()
```

```python
image_sizes = [64, 128, 256]
k_vals = list(range(3, 12, 2))

time_size_mean = []
for size in image_sizes:
    new_img = resize(img, (size, size))
    time_k = []
    for k in k_vals:
        start = time.time()
        mean_filter(new_img, k)
        end = time.time()
        time_k.append((end - start)*100)
    time_size_mean.append(time_k)

time_size_fmean = []

for size in image_sizes:
    new_img = resize(img, (size, size))
    time_k = []
    for k in k_vals:
        start = time.time()
        fast_mean_filter(new_img, k)
        end = time.time()
        time_k.append((end - start)*100)
    time_size_fmean.append(time_k)
```

```python
print(k_vals)
f = plt.figure(figsize=(20, 20))
for i, time_k in enumerate(time_size_mean):
    plt.plot(k_vals, time_k, label="Mean filter on {0}x{0}".format(image_sizes[i]))

for i, time_k in enumerate(time_size_fmean):
    plt.plot(k_vals, time_k, dashes=[6,2], label="Mean filter on {0}x{0}".format(image_sizes[i]))
    
plt.xlabel("Filter size K")
plt.ylabel("Time")
plt.legend()
plt.show()
```
