```python
from skimage.io import imread

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from Q4 import filterImg, combine_image
```

#### We now define the filters for filtering the images.

```python
# Prewitt
m1_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
m1_y = np.transpose(m1_x)

# Sobel
m2_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
m2_y = -1*np.transpose(m1_x)

# Roberts
m3_x = np.array([[ 0, 0, 0 ], [ 0, 0, 1 ], [ 0, -1, 0 ]])
m3_y = np.array([[ 0, 0, 0 ], [ 0, 1, 0 ], [ 0, 0, -1 ]])

# Laplacian
m4 = np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]])
# Laplacian with diagonals
m5 = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
```

#### Filtering the image with these filters and then combining to get resultant image.

```python
img = imread('A2_resources/barbara.jpg', as_gray=True)

prewitt = combine_image(filterImg(img, m1_x), filterImg(img, m1_y))
sobel = combine_image(filterImg(img, m2_x), filterImg(img, m2_y))
roberts = combine_image(filterImg(img, m3_x), filterImg(img, m3_y))
laplacian = filterImg(img, m4)
laplacian_diag = filterImg(img, m5)

images = [prewitt, sobel, roberts, laplacian, laplacian_diag]
titles = ["Prewitt", "Sobel", "Roberts", "Laplacian", "Laplacian with diagonals"]

f, ax = plt.subplots(3, 2, figsize=(15,15))
ax[0][0].imshow(img, 'gray')
ax[0][0].set_title("Original Image")

for i, im in enumerate(images):
    ax[(i+1)//2][(i+1)%2].imshow(images[i], 'gray')
    ax[(i+1)//2][(i+1)%2].set_title(titles[i])

f.show()
```

- Laplacian with diagonal components generates brighter images than laplacian without diagonal components. This is seen in the difference of intensity values of the two images.
