### 2: Comparison between the images obtained from modified filters and given filters.

- For filters whose sum=1,
    - When sum of the filter is made less than 1 the resultant images become dull.
    - When sum of the filter is made greater than 1 the resultant images become bright as the average value per pixel is increased.
    
- For filter whose sum=0,
    - Scaling the filter has no effect on the filter output. As the sum stays 0.
    - Filter will no longer detect edges if the sum is not 0.
    - for sum<0, it generates image negatives like those in photography.
    - for sum>0, there is no difference with the original image that is perceivable to the eye.

### 3: Why do some filters sum to 0 and others sum to 1?

- Having a filter whose coefficients sum upto 1 gives a blurring effect image. 
- If the coefficients sum to zero, the filter tries to remove low frequency components, returning only the sharp regions in the image where there is high change in the intensity values.
