```python
import numpy as np
from skimage.io import imread
import matplotlib.pyplot as plt

from Q3 import magnitude, scale_img, double_image
%matplotlib inline
```

#### We then read an image and padd it with zeros to double the size using double_image

```python
img = imread("A3_resources/64x64.png", as_gray=True)
new_img = img

for n in [64, 128, 256, 512]:
    plt.figure(figsize=(10, 10))
    plt.subplot(121)
    plt.imshow(new_img, 'gray')
    plt.title("Image of size {0}x{0}".format(n))
    plt.subplot(122)
    fft = np.fft.fft2(new_img)
    fft_shifted = np.fft.fftshift(fft)
    mag = magnitude(fft_shifted)
    scale_mag = scale_img(mag)
    plt.imshow(scale_mag, 'gray')
    plt.title("Fourier transform for {0}x{0} image".format(n))
    plt.show()

    new_img = double_image(new_img)
```

#### We observe that the FFT of the images follow the same pattern/distribution, however the resolution of the obtained FFT image increases. This is because the size of the FFT is increased but there is no new information added as all the extra padding is zero. Thus the frequencies will have the same values, but there are more frequencies in each image.
