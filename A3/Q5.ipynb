#### It is observed that the when the values of nx and ny are factors of the rows M and columns N, the sampled image is similar to the original image. The fourier transform of these images are also similar to that of the original image. We will therefore try and sample images with the sampling rates that are factors of the image shape.

```python
fft_shifted = np.fft.fftshift(np.fft.fft2(img))
mag = scale_img(magnitude(fft_shifted))

plt.figure(figsize=(10, 10))
plt.subplot(121)
plt.imshow(img, 'gray')
plt.title("Original Image".format(n))
plt.subplot(122)
plt.imshow(mag, 'gray')
plt.title("FFT of Original Image")
plt.show()

for n in [2, 4, 8, 16, 32]:
    img_new = sample_img(img, n, n)
    fft_shifted = np.fft.fftshift(np.fft.fft2(img_new))
    mag = scale_img(magnitude(fft_shifted))

    plt.figure(figsize=(10, 10))
    plt.subplot(121)
    plt.imshow(img_new, 'gray')
    plt.title("Image sampled with nx={0}, ny={0}".format(n))
    plt.subplot(122)
    plt.imshow(mag, 'gray')
    plt.title("FFT of image sampled with nx={0}, ny={0}".format(n))
    plt.show()
```

### The images have two unique frequencies and their harmonics in the fourier transforms. In samples with rate nx, ny = 2, we can see the harmonics of the frequency. We get the principal frequencies at nx=ny=16. Any further sampling would result in the loss of the prinicipal frequencies and thus the image would become unrecoverable. For nx=ny=32 the principal frequencies have been lost and the image is no longer recoverable. The same holds for nx=ny>16. This is also shown below.

```python
fft_shifted = np.fft.fftshift(np.fft.fft2(img))
mag = scale_img(magnitude(fft_shifted))

plt.figure(figsize=(10, 10))
plt.subplot(121)
plt.imshow(img, 'gray')
plt.title("Original Image".format(n))
plt.subplot(122)
plt.imshow(mag, 'gray')
plt.title("FFT of Original Image")
plt.show()

for n in range(17, 20):
    img_new = sample_img(img, n, n)
    fft_shifted = np.fft.fftshift(np.fft.fft2(img_new))
    mag = scale_img(magnitude(fft_shifted))

    plt.figure(figsize=(10, 10))
    plt.subplot(121)
    plt.imshow(img_new, 'gray')
    plt.title("Image sampled with nx={0}, ny={0}".format(n))
    plt.subplot(122)
    plt.imshow(mag, 'gray')
    plt.title("FFT of image sampled with nx={0}, ny={0}".format(n))
    plt.show()
```
