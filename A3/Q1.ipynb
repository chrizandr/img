The FFT can be found like this:


```python
import numpy as np
from skimage.io import imread
import matplotlib.pyplot as plt

from Q1 import magnitude, scale_img
%matplotlib inline
```

#### We then read an image and do the fourier transform as follows

```python
img = imread("A3_resources/5.jpeg", as_gray=True)
fft = np.fft.fft2(img)
fft_shifted = np.fft.fftshift(fft)
mag = magnitude(fft_shifted)
scale_mag = scale_img(mag)

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img, 'gray')
plt.title("Original Image")
plt.subplot(122)
plt.imshow(scale_mag, 'gray')
plt.title("Original FFT magnitude")
```

We then scale the original image and perform FFT on that to get the fourier image


```python
new_img = scale_img(img.astype(np.float))
new_fft = np.fft.fft2(new_img)
new_fft_shifted = np.fft.fftshift(new_fft)
new_mag = magnitude(new_fft_shifted)
new_scale_mag = scale_img(new_mag)

plt.figure(figsize=(10, 10))
plt.subplot(121)
plt.imshow(new_img, 'gray')
plt.title("Scaled Image")
plt.subplot(122)
plt.imshow(new_scale_mag, 'gray')
plt.title("Scaled Image FFT magnitude")
plt.show()
```

- #### Since we take a log of the intensity, the change in intensity is low.
- #### High frequency components are removed from the image and thus there is less high frequency in the FFT. Image is brighter at the center.

```python
fft_fft = np.fft.fft2(fft)
mag = magnitude(fft_fft)

plt.imshow(mag, 'gray')
plt.title("FFT of the FFT")
plt.show()
```

#### The obtained image is reversed in the x and y direction. We can flip the fourier transformed output before taking the second FFT to obtain the original image again.

```python
fft_flip = np.fft.ifftshift(np.fliplr(np.flipud(fft_shifted)))
fft_fft_flip = np.fft.fft2(fft_flip)
mag = magnitude(fft_fft_flip)

plt.imshow(mag, 'gray')
plt.title("FFT of the FFT after flipping")
plt.show()
```
