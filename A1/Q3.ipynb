### For the given five transform, in order:


#### Transform 1:
z' = 1 * z + 0.25  

for (0<=z<=0.75)
else 0


```python
img = imread("A1_resources/lena.bmp", as_gray=True)
img = img_as_float(img)

transformed = transformIntensity(img, 1, 0.25, 0, 0.75)
transformed = transformIntensity(transformed, 0, 0, 0.75, 1)
showImages(img, transformed)
```

#### Transform 2:
z' = 1.333 * z - 0.333        

for (0.25<=z<=1) 
else 0



```python
transformed = transformIntensity(img, 1.333, -0.333, 0.25, 1)
transformed = transformIntensity(transformed, 0, 0, 0, 0.25)
showImages(img, transformed)
```

#### Transform 3:
z' = 2 * z - 0.5

for (0.25<=z<=0.75)
else 0

```python
transformed = transformIntensity(img, 2, -0.5, 0.25, 0.75)
transformed = transformIntensity(transformed, 0, 0, 0, 0.25)
transformed = transformIntensity(transformed, 0, 0, 0.75, 1)
showImages(img, transformed)
```

#### Transform 4:
z' = z

for (0.25<=z<=0.75)

```python
transformed = transformIntensity(img, 1, 0, 0.25, 0.75)
transformed = transformIntensity(transformed, 0, 0, 0, 0.25)
transformed = transformIntensity(transformed, 0, 0, 0.75, 1)

showImages(img, transformed)
```
