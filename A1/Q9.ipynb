The BinarizeImage function work as follows:

```python
# Import functions to use in the code.

# Functions to read the image and convert to 8bit representation.
from skimage import img_as_ubyte
from skimage.io import imread

# Maplotlib to plot the images.
import matplotlib
%matplotlib inline

import numpy as np

from Q9 import BinarizeImage, showImages
```

Here is the working of the function

```python
# Read image and convert to 8bit grayscale
img = imread("A1_resources/puppy.jpg", as_gray=True)
img = img_as_ubyte(img)

print("Adaptive")
binary = BinarizeImage(img, "adaptive")
showImages(img, binary)

print("Global")
binary = BinarizeImage(img, "global")
showImages(img, binary)
```

#### Here is the output for palm leaf 1, with local and global thresholding:

```python
# Read image and convert to 8bit grayscale
img = imread("A1_resources/palm-leaf-1.jpg", as_gray=True)
img = img_as_ubyte(img)

print("Global Thresholding")
binary = BinarizeImage(img, "global")
showImages(img, binary)

print("Local Thresholding, with block=15, offset=5")
binary = BinarizeImage(img, "adaptive", block_size=15, offset=5)
showImages(img, binary)

print("Local Thresholding, with block=25, offset=5")
binary = BinarizeImage(img, "adaptive", block_size=25, offset=5)
showImages(img, binary)

print("Local Thresholding, with block=35, offset=5")
binary = BinarizeImage(img, "adaptive", block_size=35, offset=5)
showImages(img, binary)
```

#### Here is the output for palm leaf 1, with local and global thresholding:

```python
# Read image and convert to 8bit grayscale
img = imread("A1_resources/palm-leaf-2.jpg", as_gray=True)
img = img_as_ubyte(img)

print("Global Thresholding")
binary = BinarizeImage(img, "global")
showImages(img, binary)

print("Local Thresholding, with block=15, offset=5")
binary = BinarizeImage(img, "adaptive", block_size=15, offset=5)
showImages(img, binary)

print("Local Thresholding, with block=25, offset=5")
binary = BinarizeImage(img, "adaptive", block_size=25, offset=5)
showImages(img, binary)

print("Local Thresholding, with block=35, offset=5")
binary = BinarizeImage(img, "adaptive", block_size=35, offset=5)
showImages(img, binary)
```

##### Local Thresholding computes threshold values for each region separately. This is extremely useful for images where there are large variations of intensity values and one single threshold cannot give good enough results. It can be observed that for documents, local thresholding gives better results, but it is more expensive to compute as each region will need a separate threshold. Otsu's method computes one single threshold for the entire image.

##### In local thresholding, we must also specify the window/block size and the offset by which the block slides over the image. It can be seen that there is less noise for larger block size and more uniform intensity values for small values of offset.
