The code for the finding the k most dominant colors can be found in Q1.py file. We will import the functions defined in the file to demonstrate usbaility and give results for different questions.


The function domColors can be used as follows for both grayscale and color images:

```python
# Import functions to use in the code.

# Functions to read the image and convert to 8bit representation.
from skimage import img_as_ubyte
from skimage.io import imread

# Maplotlib to plot the images.
import matplotlib
%matplotlib inline

# Main functions for finding and showing dominant colors
from Q1 import domColors
from Q1 import showColors
```

We now plot the image and most dominant colors for a grayscale image. We will be using the **A1_resources/puppy.jpg** for this part of the code.

```python
# Read image as grayscale
img = imread("A1_resources/puppy.jpg", as_gray=True)

# Scale image to 8bit representation
img = img_as_ubyte(img)

# Find the 10 most dominant colors in the image
colors = domColors(img, 10)

# Plot the image and the colors
figure = showColors(img, colors, gray=True)
figure.show()
```

We now plot the image and most dominant colors for a color image. We will be using the A1_resources/rose.jpeg for this part of the code.

```python
# Read image as grayscale
img = imread("A1_resources/rose.jpeg")

# Scale image to 8bit representation
img = img_as_ubyte(img)

# Find the 10 most dominant colors in the image
colors = domColors(img, 10)

# Plot the image and the colors
figure = showColors(img, colors, gray=False)
figure.show()
```

### Some Uses of finding the most dominant colors in an image:

- In the Android OS, notifications for different apps are colour coded with the color of the application, these colors are not specified by the app, Android finds the most dominant colors in the application and uses those to color code the notifications.  
- In sports videos, frames can be separated into those focusing on a single player vs those focusing on the entire field by removing frames where the most dominant color is green(grass) as grass will be the most dominant in frames of the entire field.
