#### Here are the histograms for the 3 High and 3 Low contrast images:

```python
# Import necessary functions
from skimage import img_as_ubyte
from skimage.io import imread
from skimage.transform import resize

import numpy as np

import matplotlib
%matplotlib inline

from Q7 import histogram, plot_histograms
```

```python
high_contrast = ["A1_resources/hc_1.jpg", "A1_resources/hc_2.jpg", "A1_resources/hc_3.jpeg"]
low_contrast = ["A1_resources/lc_1.jpg", "A1_resources/lc_2.jpeg", "A1_resources/lc_3.jpg"]

# Read images and resize to 500x500
images = []
for img_name in high_contrast + low_contrast:
    img = imread(img_name, as_gray=True)
    img = resize(img, (500, 500), anti_aliasing=True)
    img = img_as_ubyte(img)
    images.append(img)

# Find histograms
bins = list(range(0, 257))
histograms = [histogram(x, bins) for x in images]

# Plot image and the histogram
print("High Contrast Histograms:")
for h, img in zip(histograms[0:3], images[0:3]):
    plot_histograms(h, img)

print("Low Contrast Histograms:")
for h, img in zip(histograms[3::], images[3::]):
    plot_histograms(h, img)
```

#### We can see that for High Contrast images, there are multiple peak areas in the histogram, but in general a lot of the bins are flat. In comparison, low contrast images have a much wider range of intensity values being covered.
