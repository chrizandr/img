The code for ChromaKey can be found in Q1.py file. We will import the functions defined in the file to demonstrate usbaility and give results for different questions.


The function ChromaKey can be used as follows:

```python
# Import necessary functions
from skimage import img_as_ubyte
from skimage.io import imread
from skimage.color import rgba2rgb

import numpy as np

import matplotlib
%matplotlib inline

from Q2 import resize_image
from Q2 import showImages
from Q2 import ChromaKey
```

Then read images and combine them as follows


```python

# Read the foreground and background images
fg = imread("A1_resources/fg.png", as_gray=False)
bg = imread("A1_resources/bg.png", as_gray=False)

# Resize both images to same hieght and width
fg, bg = resize_image(fg, bg)

# Convert to 8 bit RGB Image
fg = img_as_ubyte(rgba2rgb(fg))
bg = img_as_ubyte(rgba2rgb(bg))

# Define the keyColor that needs to be replaced
keyColor = np.array([0, 255, 0], dtype=np.uint8)

# Combine images
result = ChromaKey(fg, bg, keyColor, threshold=200)

# Threshold is used to specify the range of similar colored pixels that need to be replaced.
# The higher the threshold the more the range of similar colored pixels that will be replaced.
# By default threshold is set to 200, but it may need to be adjusted based on the content of the image.
# If threshold is 0, the pixels that exactly match keyColor are only replaced.

fig = showImages(fg, bg, result)
fig.show()
```

For the image in **A1_resources/rose.jpeg** I have chosen a scenic backgrounf present in **A1_resources/bg_2.jpeg**, here is the combined image for both:

```python
# Read the foreground and background images
fg = imread("A1_resources/rose.jpeg", as_gray=False)
bg = imread("A1_resources/bg_2.jpg", as_gray=False)

# Resize both images to same hieght and width
fg, bg = resize_image(fg, bg)

# Convert to 8 bit RGB Image
fg = img_as_ubyte(fg)
bg = img_as_ubyte(bg)

# Define the keyColor that needs to be replaced[black]
keyColor = np.array([0, 0, 0], dtype=np.uint8)

# Combine images
result = ChromaKey(fg, bg, keyColor, threshold=10)

fig = showImages(fg, bg, result)
fig.show()
```
