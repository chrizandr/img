The code for Q8 can be found in Q8.py file. We will import the functions defined in the file to demonstrate usbaility and give results for different questions.


The BitSlice function work as follows:

```python
# Import functions to use in the code.

# Functions to read the image and convert to 8bit representation.
from skimage import img_as_ubyte
from skimage.io import imread

# Maplotlib to plot the images.
import matplotlib
%matplotlib inline

import numpy as np

from Q8 import BitSlice, showSlice
```

Here is the usage of the function


```python
img = imread("A1_resources/puppy.jpg", as_gray=True)
img = img_as_ubyte(img)

bitslice = BitSlice(img, 5)
showSlice(bitslice)
```

Here are all 8 bit slices of the image.

```python
for i in range(0, 8):
    print("{} bit slice:".format(i))
    bitslice = BitSlice(img, i)
    showSlice(bitslice)
```

Similarly for the lena image:

```python
img = imread("A1_resources/lena.bmp", as_gray=True)
img = img_as_ubyte(img)

for i in range(0, 8):
    print("{} bit slice:".format(i))
    bitslice = BitSlice(img, i)
    showSlice(bitslice)
```

#### We observe that the higher the bit which is sliced, the more information is contained in the image. Therefore, we can conclude that the 7th bit slices contain the most similarity with the original images.
